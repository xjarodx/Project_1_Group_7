```python
# Dependencies
import numpy as np
import pandas as pd
import requests
import json
import urllib
import time
```

```python
# START HERE!!!
firefile = '122000.json'
with open(firefile) as jsonfile:
    fire_data = json.load(jsonfile)
    
fire_data['features'][0]['attributes']
```

```python
#Once we have all the json information we can parse them out as such
incidents = []
lat = []
lon = []
# loop through each set and parse the data

for i in fire_data['features']:
    try:
        incidents.append(i['attributes'])
        lat.append(i['geometry']['y'])
        lon.append(i['geometry']['x'])
    except:
        lat.append('missing')
        lon.append('missing')
#             print('Missing Geometry... ')
len(incidents)
```

```python
df = pd.DataFrame(incidents)
```

```python
df.head()
```

```python
df['Longitude'] = lon
df['Latitude'] = lat
```

```python
df.head()
```

```python
df.to_csv("output/extra_raw_fire_data.csv", index=False, header=True, float_format='%.f')
```
