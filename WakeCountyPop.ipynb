```python
#import dependencies
import requests
import pandas as pd

#Get Wake County Population
wake_county_pop_query_url = "https://api.opendatanetwork.com/data/v1/values?app_token=cQovpGcdUT1CSzgYk0KPYdAI0&describe=true&forecast=5&format=google&variable=demographics.population.count&entity_id=0500000US37183"
wake_county_pop_response = requests.get(wake_county_pop_query_url)
wake_county_pop_json = wake_county_pop_response.json()
```

```python
# Create an empty data frame for Wake County Population
column_names = ("Year", "Population")
wakecountypop_df = pd.DataFrame(columns = column_names)
wakecountypop_df
```

```python
wake_county_pop_year = []
wake_county_pop_count = []
for i in wake_county_pop_json["data"]["rows"]:

    wake_county_pop_year.append(i["c"][0]["v"])
    wake_county_pop_count.append(i["c"][1]["v"])
    
wakecountypop_df["Year"] = wake_county_pop_year
wakecountypop_df["Population"] = wake_county_pop_count
wakecountypop_df
```
